# Processing the raw data file, do it only once !!!!!!

```python
cell_id_list =['20240826001',
              '20240826002',
               '20240826003',
               '20240826004',
               '20240826005',
               '20240826006'] 

for cell_id in cell_id_list:
    filename =cell_id+ '.xlsx'
    cycle = pd.read_excel(filename,sheet_name='cycle')
    step = pd.read_excel(filename,sheet_name='step')
    details = pd.read_excel(filename,sheet_name='record')

    cycle.to_csv(cell_id+'_cycle.csv')
    step.to_csv(cell_id+'_step.csv')
    details.to_csv(cell_id+'_details.csv')
```

## obtain the C/20 Charge discharge capacity

```python
df_step6
```

```python
list_cap_c20_char = []
list_cap_c20_dchar = []
list_cap_c10_dchar = []
list_cap_c5_dchar = []
list_cap_c2_dchar = []
list_cap_1c_dchar = []
list_cap_2c_dchar = []

list_k_val = []

datalist = [df_step1,df_step2,df_step3,df_step4,df_step5,df_step6]

for df in datalist:
    #C/20 charge capacity
    cap_c20_char = df.loc[df['Step Number']==6]['Capacity(Ah)'].values[0]
    list_cap_c20_char.append(cap_c20_char)
    
    #C/20 discharge capacity
    cap_c20_dchar = df.loc[df['Step Number']==8]['Capacity(Ah)'].values[0]
    list_cap_c20_dchar.append(cap_c20_dchar)
    
    # K value
    k = (df.loc[df['Step Number']==6]['End Voltage(V)'].values[0] - df.loc[df['Step Number']==7]['End Voltage(V)'].values[0])/12
    list_k_val.append(k)
    
    # C rate performance
    cap_c10_dchar = df.loc[df['Step Number']==12]['Capacity(Ah)'].values[0]
    list_cap_c10_dchar.append(cap_c10_dchar)
    
    cap_c5_dchar = df.loc[df['Step Number']==16]['Capacity(Ah)'].values[0]
    list_cap_c5_dchar.append(cap_c5_dchar)
    
    cap_c2_dchar = df.loc[df['Step Number']==20]['Capacity(Ah)'].values[0]
    list_cap_c2_dchar.append(cap_c2_dchar)
    
    cap_1c_dchar = df.loc[df['Step Number']==24]['Capacity(Ah)'].values[0]
    list_cap_1c_dchar.append(cap_1c_dchar)
    
    cap_2c_dchar = df.loc[df['Step Number']==28]['Capacity(Ah)'].values[0]
    list_cap_2c_dchar.append(cap_2c_dchar)

k_val_12hr = [x*1000/12 for x in list_k_val]
    
c_20_ce = [x/y*100 for x,y in zip(list_cap_c20_dchar,list_cap_c20_char)]
list_cap_c10_dchar_ref =  [x/y*100 for x,y in zip(list_cap_c10_dchar, list_cap_c20_dchar)]
list_cap_c5_dchar_ref =  [x/y*100 for x,y in zip(list_cap_c5_dchar, list_cap_c20_dchar)]
list_cap_c2_dchar_ref =  [x/y*100 for x,y in zip(list_cap_c2_dchar, list_cap_c20_dchar)]
list_cap_1c_dchar_ref =  [x/y*100 for x,y in zip(list_cap_1c_dchar, list_cap_c20_dchar)]
list_cap_2c_dchar_ref =  [x/y*100 for x,y in zip(list_cap_2c_dchar, list_cap_c20_dchar)]
```

```python
for data in [list_cap_c20_char,list_cap_c20_dchar,c_20_ce,k_val_12hr,list_cap_c10_dchar_ref,list_cap_c5_dchar_ref,list_cap_c2_dchar_ref,list_cap_1c_dchar_ref,list_cap_2c_dchar_ref]:
    print ('mean',np.mean(data), 'std',np.std(data))
```
